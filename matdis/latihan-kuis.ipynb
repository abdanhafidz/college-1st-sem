```python
def nigga(n):
        if(n == 0):
             return 0
        else:
            if(n%2 == 1):
                return 1 + nigga(n//2)
            else:
                return nigga(n//2)
print(nigga(5))
```

nigga(7) = 1 + nigga(3) = 1 + 2 = 3
