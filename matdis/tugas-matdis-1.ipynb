<p>
a) p → (¬q ∨ r) <br>
b) ¬p → (q → r) <br>
c) (p → q) ∨ (¬p → r) <br>
d) (p → q) ∧ (¬p → r) <br>
e) (p ↔ q) ∨ (¬q ↔ r) <br>
f) (¬p ↔ ¬q) ↔ (q ↔ r) </p>

```python
def Biimply(p,q):
    return ((q if p else True) and (p if q else True)) 


def kerjain(no,p,q,r):
    match no:
        case 1:
            return (not q or r) if p else True
        case 2:
            return (r if q else True) if not p else True
        case 3:
            return (q if p else True) or (r if not p else True)
        case 4:
            return (q if p else True) and (r if not p else True)
        case 5:
            return Biimply(p,q) or Biimply(not q, r)
        case 6:
            return Biimply(Biimply(not p,not q),Biimply(q, r))

def kerjainSoal(opr,no):
    noSoal = ["a","b","c","d","e","f"]
    print("Soal",noSoal[no-1],"). ")
    print("p | q | r | ",opr)
    print("-" * 23)

    for p in [True, False]:
        for q in [True, False]:
            for r in [True, False]:
                result = kerjain(no,p, q, r)
                print(f"{int(p)} | {int(q)} | {int(r)} | {int(result)}")
                
    print("\n")
soal = ["p → (¬q ∨ r)","¬p → (q → r)","(p → q) ∨ (¬p → r)","(p → q) ^ (¬p → r)","(p ↔ q) ∨ (¬q ↔ r)","(¬p ↔ ¬q) ↔ (q ↔ r)"]
i = int(1)
for x in soal:
    kerjainSoal(x,i)
    i+=1
```

logically equivalent.
