```python

def insertion_sort(a):
    step = 0
    for i in range(1, len(a)):
        j = i - 1
        key = a[i]
        while (a[j] > key) and (j >= 0):
            print("Step ke-",step)
            print("Sebelum:",a)
            step += 1
            print(a[j+1],"<->",a[j])
            a[j + 1] = a[j]
            j -= 1
            a[j + 1] = key
            print("Sesudah:",a)
    print("banyak step untuk insertion sort adalah =", step)

def bubble_sort(a):
    step = 0
    n = len(a)
    for i in range(0, n):
        for j in range(0, n - i - 1):
            if a[j] > a[j+1]:
                print("Step ke-",step)
                print("sebelum",a)
                step += 1
                temp = a[j]
                print(a[j+1],"<->",a[j])
                a[j] = a[j+1]
                print("sesudah",a)
                a[j+1] = temp


    print("banyak step adalah bubble sort =", step)
```

```python
from math import floor
def BinarySearch(a, t):
    found = False
    L = 0
    R = len(a) - 1
    step = 0
    while L <= R:
        m = floor((L + R) / 2)
        step+=1
        if a[m] < t:
            L = m + 1
        elif a[m] > t:
            R = m - 1
        else:
            found = True
            break
    if(found):
        print(t,"is at index", m)
    else:
        print(t,"not found")
    print("banyak step adalah =", step)


```

```python
procedure initialize_memo()
    memo = array of size n
    for i = 1 to n do
        memo[i] = -1
    end for
    memo[1] = 1
    memo[2] = 1

function f(n)
    if n <= 0 then
        return 1
    else if memo[n] != -1 then
        return memo[n]
    else
        memo[n] = f(n-1) + 2*f(n-2)
        return memo[n]
    end if
end function

initialize_memo()
print(f(n))
```
